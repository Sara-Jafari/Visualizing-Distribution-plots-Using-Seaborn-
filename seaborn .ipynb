```python
#importing the seaborn
import seaborn as sns
#setting the display within the jupyter notebook 
%matplotlib inline
import warnings

#Ignore warnings
warnings.filterwarnings("ignore")
```

```python
#loading the dataset
planets = sns.load_dataset("planets")
planets.head()
```

```python
planets.columns
```

```python
planets.shape
```

```python
planets.info()
```

```python
#The dataset has 6 columns and 1,035 rows. It also contains both categorical and numerical columns.
```

```python
# We will also check whether there are any missing values in the dataset.
planets.isna().sum()
```

```python
planets.dropna(inplace=True)
```

```python
planets.shape
```

```python
#distplot
sns.distplot(planets ["mass"])
```

```python
sns.distplot(planets["mass"] , kde = False)
```

```python
#jointplot
sns.jointplot(x ="year" , y="distance" , data = planets , kind = "scatter")
```

```python
sns.jointplot(x ="year" , y="distance" , data = planets , kind = "hex")
```

```python
sns.jointplot(x ="year" , y="distance" , data = planets , kind = "reg")
```

```python
sns.jointplot(x ="year" , y="distance" , data = planets , kind = "kde")
```

```python
#pairplot
sns.pairplot(planets)
```

```python
#Specify the color of batch data and palette (color scheme)
sns.pairplot(planets , hue="method" , palette='Blues')
```

```python
#Rug plot
sns.rugplot(planets["distance"])
```

```python
#kde plot
sns.kdeplot(planets["distance"])
sns.rugplot(planets["distance"])
```
